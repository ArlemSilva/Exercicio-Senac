```python
import pandas as pd
```

```python
df1 = pd.read_csv('ca-2020-01.csv', sep = ';')
df2 = pd.read_csv('ca-2020-02.csv', sep = ';')
df3 = pd.read_csv('ca-2021-01.csv', sep = ';')
df4 = pd.read_csv('ca-2021-02.csv', sep = ';', encoding='latin1')
df5 = pd.read_csv('ca-2022-01.csv', sep = ';')
df6 = pd.read_csv('ca-2022-02.csv', sep = ';')
df7 = pd.read_csv('ca-2023.01.csv', sep = ';')
df8 = pd.read_csv('ca-2023.02.csv', sep = ';')
df9 = pd.read_csv('ca-2024.01.csv', sep = ';')
df10 = pd.read_csv('ca-2024.02.csv', sep = ';')
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10])
```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float) # trocando as virgulas (,) dos valores por ponto (.) & forçando os valores como float com a sintaxe .astype(float)
print(df['Valor de Venda'].dtype) # .dtype para mostrar que tipo de formato é o dado
```
